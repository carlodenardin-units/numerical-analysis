# Interpolation
Given n + 1 points $\{X\}^{n}_{i = 0}$ in the interval $[0, 1]$, the lagrange interpolation operator is: 
$$L^{n}: C^{0}([0, 1]) \rightarrow P^{n}$$
where $C^{0}([0, 1])$ is an infinitely dimensional space and $P^{n}$ is a finite dimensional space of polynomials of order n
such that:
$$(L^{n}f)(X_{i}) = f(X_{i}), i = 0, ..., n$$

## Monomial basis
The finite dimensional space $P^{n}$ has dimension $n + 1$ and can be constructed using $n + 1$ linear combinations of linear independent polynomials of order $\le n$, for example the monomials:
$$P^{n} = span\{v_{i} := x^{i}\}^{n}_{i = 0}$$

NB: If we want to constuct the Lagrange interpolation of a given function on equispaced $n + 1$ in $[0, 1]$, then we are actively looking for an element of $P^{n}$ that coincides with the function at there given points.

## Problem formulation
Given a basis $\{v_{i}\}^{n}_{i = 0}$ any element of $P^{n}$ can be written as a linear combination of the basis, i.e.,
$$\forall p \in P^{n}, \exists ! \{p^{i}\}^{n}_{i = 0} | p(x) = \sum^{n}_{i = 0} p^{i}v_{i}(x)$$
If we want to solve the interpolation problem above, then we need to find the coefficients $p^i$ of the polynomial $p$ that interpolates at the points $X_i$:
$$v_{j}(X_i)p^{j} = f(X_i) \iff [[V]][p] = [F]$$
At the end interpolation means solving the system $$[[V]][p] = [F]$$

## Lebesgue constant
Recall the inequality and the Lebesgue function:
$$||L^{n}u||_{L^{\infty}} \le ||V^{-1}||_{l^{\infty}} ||\wedge||_{L^{\infty}} ||u||_{L^{\infty}}$$
where: $\wedge = \sum_{j}|v_{j}|$

The best choice is to have the norm of the basis equal to 1. To obtain a good approximation we need to have many info but if we have many info we may end up in a bad conditioned problem. This is a common problem with the monomial basis since is we have $n = 1000$ and $m = 1000$ they may not linear independent.

## Lagrange basis
Given n + 1 points $\{X\}^{n}_{i = 0}$ in the interval $[0, 1]$, the lagrange interpolation operator is: 
$$L^{n}: C^{0}([0, 1]) \rightarrow P^{n}$$
where $C^{0}([0, 1])$ is an infinitely dimensional space and $P^{n}$ is a finite dimensional space of polynomials of order n
such that:
$$(L^{n}f)(x) = \sum^{n}_{i = 0} f(X_{i})l_{i}(x), i = 0, ..., n$$
where:
$$l_{i}(x) := \prod^{n}_{i \ne j, j = 0} \frac{x - x_{j}}{x_{i} - x_{j}}$$

It is used the lagrange basis as the matrix $V$ is the identity matrix $V^{-1}$ so it is 1.

## Monomial Interpolation Example
Below an example of the monomial interpolation is reported. We want to interpolate the absolute function given $n + 1$ interpolation points.

```python
%pylab inline
```

```python
from interpolation import Monomial

# Define the function to approximate
def absoluteFunction(x):
    return abs(x - 0.5)

func = absoluteFunction # Function to approximate
x = linspace(0, 1, 100) # Points used to plot the function
X = linspace(0, 1, 10) # Points used to approximate the function

# Compute the interpolation with the monomial basis
monomial = Monomial()
B, V, u, p = monomial.computeInterpolation(func = func, x = x, X = X)

plot(x, func(x), label = 'Function to approximate')
plot(X, func(X), 'go', label = 'Interpolation points')
plot(x, B.dot(p), label = 'Approximation')
legend()
title('Monomial Interpolation')
show()
```
