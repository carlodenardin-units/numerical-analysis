# Matplotlib
With `matplotlib` is possible to manage all the parts of a graph. In this image are reported all the parts of the figure that we can manage and modify.

<img src="./images/matplotlib%20figure%20parts.png" alt="drawing" width="400"/>

```python
import matplotlib.pyplot as plt
import numpy as np

linear = np.linspace(-1, 1, 25) # Return linear spaced values from 0 to 2
cubic = linear ** 3
plt.plot(linear, linear, 'o-', label = 'linear')
plt.plot(linear, cubic, 'go', label = 'cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.grid(linestyle = 'dotted')
plt.show()
# plt.savefig('myploy.pdf')
```

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'o-', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2), 'g--')
plt.show()
```
