```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_ = np.array([1, 3, 4, 6, 10])
y_ = np.array([0, 2, 7, 20, 54])

n = len(x_)
B = np.matrix([np.ones(n), x_, x_**2, x_**3, x_**4]).T
BT = B.T

y = np.matrix(y_).T

a = np.linalg.solve(BT*B,BT*y)

a0 = float(a[0])
a1 = float(a[1])
a2 = float(a[2])
a3 = float(a[3])
a4 = float(a[4])

xx = np.linspace(0,10,100)
f = lambda x: a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4

plt.plot(xx,f(xx))

plt.scatter(x_, y_, color = 'r', marker = 'o')
```

```python
year = np.array([1900, 1910, 1920, 1930, 1941, 1950, 1960, 1970, 1980, 1990, 2000, 2010])
population = np.array([3315, 3753, 3880, 4066, 4266, 4715, 5429, 6270, 6366, 6874, 7288, 7783])

n = len(year)
B = np.matrix([np.ones(n), year, year**2]).T
BT = B.T

y = np.matrix(population).T

a = np.linalg.solve(BT*B,BT*y)

a0 = float(a[0])
a1 = float(a[1])
a2 = float(a[2])

xx = np.linspace(1900,2010,100)
f = lambda x: a0 + a1*x + a2*x**2

plt.plot(xx,f(xx))
plt.scatter(year,population, color= 'r',marker='o')

print(f(1945))
print(f(1975))
print(f(2020))
```
